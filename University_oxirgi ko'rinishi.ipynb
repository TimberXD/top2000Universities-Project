```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

topunivers = pd.read_csv("topuniversities.csv", encoding=('ISO-8859-1'))
topunivers
```

# Top 8 talik Universitetlarda alumni employment darajalari


```python
dataset_univ = topunivers.sort_values(by=['world_rank'],ascending=True)
sakkiz_univ = dataset_univ.head(8)

sorted_dataset = sakkiz_univ.sort_values(by=["alumni_employment"],ascending=True)
sorted_dataset

sorted_alumni = sorted_dataset["alumni_employment"]
sorted_alumni

sorted_nomi = sorted_dataset['institution']
sorted_nomi


plt.scatter(sorted_alumni, sorted_nomi)
plt.title("Top 8 talik Universitetlarda alumni employment darajalari")
plt.xlabel("Alumni employment bo'yicha o'rni")
plt.ylabel("Universitet nomlari")

plt.show()
```

# Amerikaning top 25 universitetlarining xalqaro reytingda darajalari (BARH)

```python
amerika_univ = topunivers.groupby(['country']).get_group("USA")
firstamerika_grouped = amerika_univ.head(25)
firstamerika_grouped 

univ_nametion = firstamerika_grouped.institution
univ_score = firstamerika_grouped.score

plt.title("Amerikaning top 25 universitetlarining xalqaro reytingda darajalari")
plt.xlabel("Xalqaro reyting darajasi")
plt.ylabel("Universitet nomlari")

plt.barh(univ_nametion,univ_score,)
```

# United Kingdomning eng past va eng yuqori ballarga ega universitetlari top 5talik (step)

```python
uk_univ = topunivers.groupby(['country']).get_group("United Kingdom")
uk_grouped = uk_univ.head(10)
uk_second = uk_univ.tail(10)

uk_nametion = uk_grouped.institution
uk_score = uk_grouped.score

second_name = uk_second.institution
second_score = uk_second.score

plt.title("United Kingdomning eng past va eng yuqori ballarga ega universitetlari top 5talik")
plt.xlabel("Xalqaro ballar darajasi")
plt.ylabel("Universitet nomlari")

plt.step(uk_score,uk_nametion)
plt.step(second_score,second_name)
```

# Germaniyaning top 20 universitetlaridagi o'quv sifati darajalari
